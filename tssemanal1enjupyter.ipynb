```python
import numpy as np

def mi_funcion_sen(vmax, dc, ff, phi, nn, fs):
    """

    """
    tt = np.arange(0, nn/fs,1/fs)  # Vector de tiempo
    xx = vmax * np.sin(2 * np.pi * ff * tt + phi) + dc  # Se√±al senoidal

    return tt, xx

def mi_cuadrada(amp, freq, duty, Nsamples, freqs):

    ttt = np.arange(0, Nsamples/freqs, 1/freqs)
    xxx = np.zeros_like(ttt)

    period = 1 / freq
    estado_alto = duty * period

    for i in range(Nsamples):
        if (ttt[i] % period) <  estado_alto:
            xxx[i] = amp
        else:
            xxx[i] = -amp

    return ttt, xxx
```

```python
N = 10  # Num de muestras
ff = 1000 #Frec a muestrear
fs = 10000  # Frec de muestreo en Hz
vmax=1
dc=0
phi=0 # en rad

tt, xx = mi_funcion_sen(vmax=vmax, dc=dc, ff=ff, phi=phi, nn=N, fs=fs)

# Para visualizar la senoidal:
import matplotlib.pyplot as plt
plt.plot(tt, xx)
plt.title('Se√±al Senoidal')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')
plt.grid(True)
plt.show()

#################################3
amp= 1
freq=1000
duty=0.5
Nsamples=20
freqs=10000


ttt, xxx = mi_cuadrada(amp=amp, freq=freq, duty=duty, Nsamples=Nsamples, freqs=freqs)

# Para visualizar la se√±al senoidal:
import matplotlib.pyplot as plt
plt.plot(ttt, xxx)
plt.title('Se√±al cuadrada')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')
plt.grid(True)
plt.show()
```
